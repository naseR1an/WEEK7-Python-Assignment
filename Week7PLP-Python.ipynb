```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species column
df["species"] = iris.target_names[iris.target]

# Save to CSV for demonstration
df.to_csv("iris_data.csv", index=False)
```

```python
print(df.head())  # Displays the first five rows
```

```python
print(df.info())  # Check data types
print(df.isnull().sum())  # Check for missing values
```

```python
print(df.describe())  # Summary statistics for numerical columns
```

```python
print(df.groupby("species").mean())  # Compute mean values per species
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fake time-series data (for example, sales trends)
df["date"] = pd.date_range(start="2025-01-01", periods=len(df), freq="D")
df["sales"] = np.random.randint(50, 200, size=len(df))

plt.figure(figsize=(8,5))
plt.plot(df["date"], df["sales"], marker="o", color="blue")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Trends Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.barplot(x="species", y="sepal length (cm)", data=df, palette="pastel")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.title("Sepal Length Across Iris Species")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["sepal width (cm)"], bins=10, color="skyblue", edgecolor="black")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Sepal Length vs Petal Length")
plt.show()
```
